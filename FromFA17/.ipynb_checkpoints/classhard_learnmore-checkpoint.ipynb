```python
import pandas as pd
```

```python
df = pd.read_csv("realdata0.csv")
```

```python
for i in range(20):
    print(i)
    df = df.merge(pd.read_csv("realdata"+ str(i)+".csv", low_memory=False), how="outer")
```

```python
df
```

```python
data = df[(df.Question == "I learned a great deal from this course.") | 
   (df.Question =="This course is difficult relative to others.")]
```

```python
data
```

```python
data.iloc[0].to_dict()
```

```python
data.columns
```

```python
data = data[["term", "course", "Question",'Strongly Disagree', 'Disagree',
       'Neither Agree nor Disagree', 'Agree', 'Strongly Agree']]
```

```python
data["w_sum"] = (5*data["Strongly Agree"] + 4*data["Agree"] + 3*data["Neither Agree nor Disagree"] + 
                 2*data["Disagree"] + data["Strongly Disagree"])
```

```python
data
```

```python
data["mean"] = data.w_sum/ data[['Strongly Disagree', 'Disagree',
       'Neither Agree nor Disagree', 'Agree', 'Strongly Agree']].sum(axis = 1)
```

```python
data
```

```python
x = data[data.Question == "I learned a great deal from this course."].reset_index()
x
```

```python
x[x["mean"].isna()]
```

```python
y = data[data.Question == "This course is difficult relative to others."].reset_index()
y
```

```python
import matplotlib.pyplot as plt
```

```python
xy = x.merge(y, on=["term", "course"])
xy.plot.scatter(x="mean_x", y="mean_y")
```
