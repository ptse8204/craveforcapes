# Cleaning
Get rid everything that is Fa17, since it is only paritally collected

```python
data = df[df["term"] != "FA17"]
data
```

In this analysis, we are just focusing on real scores, thus, we would just analyze `Grade received`.

```python
data = data[data.Question == "Grade received"]
data
```

```python
data.columns
```

Since we are only checking grades, we only need course, term, and grade A-F and P/NP

```python
data = data[["term", "course", 'A', 'B', 'C',
       'D', 'F', 'P', 'NP']]
data
```

```python
data.isna().any()
```

Drop na here, since we are sure that the nan data only exist in the grade data, thus we could drop any na. Since these are corrupted.

```python
data = data.dropna()
data
```

Using the % value is slightly easier, so we would only need the % values.

```python
data = data[data.A.str.contains("%")]
data
```

Let's strip the % so that it would be a int

```python
data.columns
```

```python
for column in data.columns[2:]:
    data[column] = data[column].str.replace("%", "").astype("int")
```

```python
data
```

Let's drop summer quarters, because it is not that useful

```python
data = data[data["term"].str.contains("SP|WI|FA")]
data
```

# Second Analysis
Check the percentage of CAPE received

```python
cape_return = df[df.Question == "General Info"].dropna(axis=1)
cape_return
```

```python
cape_return["fraction"] = cape_return["CAPEs Returned"]/cape_return["Enrollment"]
cape_return
```

```python
cape_return = cape_return[["term", "course", "fraction", "Enrollment", "CAPEs Returned"]]
```

```python
cape_return[cape_return.term != "FA17"]
```

```python
cape_return.fraction.hist()
```

Upon further inspection, it seems that these data likely has the enrollment and returned flipped. Let's flip it back.

```python
cape_return[cape_return.fraction > 1]
```

```python
for k in cape_return.iterrows():
    i = k[1]
    if int(i["Enrollment"]) < int(i["CAPEs Returned"]):
        real_enroll = i["CAPEs Returned"]
        real_return = i["Enrollment"]
        i["Enrollment"] = real_enroll
        i["CAPEs Returned"] = real_return
        i.fraction = int(i["CAPEs Returned"])/int(i["Enrollment"])
```

```python
cape_return[cape_return.fraction > 1]
```
